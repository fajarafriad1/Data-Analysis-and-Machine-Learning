```python
import pandas as pd
import os
import warnings
import requests
from bs4 import BeautifulSoup as bs
import time
import re
warnings.filterwarnings("ignore")
from google.colab import files
```

```python
!pip install selenium
!pip install webdriver_manager
!apt-get update
!apt-get install -y chromium-browser
!apt-get install chromium-chromedriver

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.common.action_chains import ActionChains
```

### Initiating Selenium

```python
options = webdriver.ChromeOptions()
options.add_argument("user-agent=whatever you want")
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_experimental_option("detach", True)
options.add_experimental_option("excludeSwitches", ["enable-automation"])
```

### Go to the website

```python
wd = webdriver.Chrome(options=options)
url = "https://www.mobil123.com/mobil-dijual/indonesia"            #url
wd.get(url)
WebDriverWait(wd, 5)
time.sleep(5)
```

### Collecting Product Link

```python
link_product = []
page = 1
product = 1
while page < 4:
  #print('page', page)
  #print(wd.current_url)
  num_of_products = len(link_product)
  #print('num of products:', num_of_products)
  html = wd.page_source
  soup = bs(html)
  #print(soup)
  h2 = soup.find_all('h2', class_='listing__title epsilon flush')
  for info in h2:
    link = info.find('a',href=True)
    link = link['href']
    link_product.append(link)
    print('produk', product)
    product += 1

  WebDriverWait(wd, 5).until(EC.element_to_be_clickable((By.XPATH, "//span[text()='Selanjutnya']"))).click()
  time.sleep(5)
  print('page', page)
  page += 1
len(link_product)
```

```python
col_name = ['merk', 'tipe', 'varian', 'harga', 'kondisi', 'tahun_kendaraan', 'kilometer', 'warna', 'mesin', 'transmisi', 'penumpang', 'pintu', 'dirakit', 'tenaga_puncak_hp', 'tenaga_putaran_puncak_nm', 'injeksi_langsung', 'aspirasi', 'bahan_bakar', 'panjang_mm', 'lebar_mm', 'tinggi_mm', 'jarak_roda_mm', 'seller_name', 'seller_status', 'seller_lokasi', 'judul', 'tanggal']
list_mobil = []
```

#### Collecting Product Data

```python
for i in range(len(link_product)):
  wd = webdriver.Chrome(options=options)
  print('product:', i+1)
  wd.get(link_product[i])
  time.sleep(2)
  #print(wd.current_url)
  product_html = wd.page_source
  product_soup = bs(product_html)
  ##print(product_soup)

  product_name = product_soup.find_all('a', itemprop="item")

  try:
    merk = product_name[2].find('span')
    merk = merk.text.strip() if merk != None else ''
    #print('merk', merk)
  except: merk = ''
  try:
    tipe = product_name[3].find('span')
    tipe = tipe.text.strip() if tipe != None else ''
    #print('tipe', tipe)
  except: tipe = ''
  try:
    varian = product_name[4].find('span')
    varian = varian.text.strip() if varian != None else ''
    #print('varian', varian)
  except: varian = ''
  try:
    harga = product_soup.find('h3', class_='u-color-white u-text-3 u-text-4@mobile u-text-bold u-margin-bottom-none u-margin-top-xxs')
    harga = harga.text.strip() if harga != None else ''
    #print('harga', harga)
  except: harga = ''

  general_info = product_soup.find_all('div', class_='c-card__body u-bg-haze-light u-flex u-flex--column')

  try:
    kondisi = general_info[0].find('span', class_='u-text-bold u-block')
    kondisi = kondisi.text.strip() if kondisi != None else ''
    #print('kondisi', kondisi)
  except: kondisi = ''
  try:
    tahun_kendaraan = general_info[1].find('span', class_='u-text-bold u-block')
    tahun_kendaraan = tahun_kendaraan.text.strip() if tahun_kendaraan != None else ''
    #print('tahun_kendaraan', tahun_kendaraan)
  except: tahun_kendaraan = ''
  try:
    kilometer = general_info[2].find('span', class_='u-text-bold u-block')
    kilometer = kilometer.text.strip() if kilometer != None else ''
    #print('kilometer', kilometer)
  except: kilometer = ''
  try:
    warna = general_info[3].find('span', class_='u-text-bold u-block')
    warna = warna.text.strip() if warna != None else ''
    #print('warna', warna)
  except: warna = ''
  try:
    mesin = general_info[4].find('span', class_='u-text-bold u-block')
    mesin = mesin.text.strip() if mesin != None else ''
    #print('mesin', mesin)
  except: mesin = ''
  try:
    transmisi = general_info[5].find('span', class_='u-text-bold u-block')
    transmisi = transmisi.text.strip() if transmisi != None else ''
    #print('transmisi', transmisi)
  except: transmisi = ''
  try:
    penumpang = general_info[6].find('span', class_='u-text-bold u-block')
    penumpang = penumpang.text.strip() if penumpang != None else ''
    #print('penumpang', penumpang)
  except: penumpang = ''
  spesifikasi = product_soup.find_all('div', class_='o-grid o-grid--lg u-margin-ends-lg')
  try:
    jumlah_gir = spesifikasi[0].find(text=re.compile('Jumlah Gir'))
    jumlah_gir = jumlah_gir.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('jumlah_gir', jumlah_gir)
  except: jumlah_gir = ''
  try:
    pintu = spesifikasi[1].find(text=re.compile('Pintu'))
    pintu = pintu.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('pintu', pintu)
  except: pintu = ''
  try:
    dirakit = spesifikasi[1].find(text=re.compile('Dirakit'))
    dirakit = dirakit.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('dirakit', dirakit)
  except: dirakit = ''
  try:
    baut_mm = spesifikasi[2].find(text=re.compile('Baut'))
    baut_mm = baut_mm.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('baut_mm', baut_mm)
  except: baut_mm = ''
  try:
    struk_mm = spesifikasi[2].find(text=re.compile('Struk'))
    struk_mm = struk_mm.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('struk_mm', struk_mm)
  except: struk_mm = ''
  try:
    tenaga_puncak_hp = spesifikasi[2].find(text=re.compile('Tenaga Puncak'))
    tenaga_puncak_hp = tenaga_puncak_hp.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('tenaga_puncak_hp', tenaga_puncak_hp)
  except: tenaga_puncak_hp = ''
  try:
    tenaga_putaran_puncak_nm = spesifikasi[2].find(text=re.compile('Tenaga Putaran Puncak'))
    tenaga_putaran_puncak_nm = tenaga_putaran_puncak_nm.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('tenaga_putaran_puncak_nm', tenaga_putaran_puncak_nm)
  except: tenaga_putaran_puncak_nm = ''
  try:
    injeksi_langsung = spesifikasi[2].find(text=re.compile('Injeksi Langsung'))
    injeksi_langsung = injeksi_langsung.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('injeksi_langsung', injeksi_langsung)
  except: injeksi_langsung = ''
  try:
    aspirasi = spesifikasi[2].find(text=re.compile('Aspirasi'))
    aspirasi = aspirasi.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('aspirasi', aspirasi)
  except: aspirasi = ''
  try:
    bahan_bakar = spesifikasi[2].find(text=re.compile('Tipe Bahan Bakar'))
    bahan_bakar = bahan_bakar.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('bahan_bakar', bahan_bakar)
  except: bahan_bakar = ''
  try:
    panjang_mm = spesifikasi[3].find(text=re.compile('Panjang'))
    panjang_mm = panjang_mm.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('panjang_mm', panjang_mm)
  except: panjang_mm = ''
  try:
    lebar_mm = spesifikasi[3].find(text=re.compile('Lebar'))
    lebar_mm = lebar_mm.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('lebar_mm', lebar_mm)
  except: lebar_mm = ''
  try:
    tinggi_mm = spesifikasi[3].find(text=re.compile('Tinggi'))
    tinggi_mm = tinggi_mm.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('tinggi_mm', tinggi_mm)
  except: jarak_roda_mm = ''
  try:
    jarak_roda_mm = spesifikasi[3].find(text=re.compile('Jarak Roda'))
    jarak_roda_mm = jarak_roda_mm.find_next("span", class_='u-text-bold u-width-1/2 u-align-right').get_text(strip=True)
    #print('jarak_roda_mm', jarak_roda_mm)
  except: jarak_roda_mm = ''

  seller = product_soup.find('div', class_='c-card c-card--ctr u-margin-ends-sm u-order-last@mobile')

  try:
    seller_name = seller.find('h6').text.strip()
    #print(seller_name)
  except: seller_name = ''

  try:
    seller_info = seller.find('div', class_='u-flex u-align-items-center u-flex--wrap').find_all('span')[0]
    seller_status = seller_info.text.strip()
  except: seller_status = ''

  try:
    seller_info = seller.find('div', class_='u-flex u-align-items-center u-flex--wrap').find_all('span')[2]
    seller_lokasi = seller_info.text.strip()
  except: seller_status = ''

  try:
    judul = product_soup.find('h1', class_='listing__title u-color-dark u-margin-bottom-xs u-text-3 u-text-4@mobile u-text-bold')
    judul = judul.text.strip() if judul != None else ''
    #print('judul', judul)
  except: judul = ''
  try:
    tanggal = product_soup.find('span', class_='u-color-muted u-text-7 u-hide@desktop u-margin-bottom-xs u-block')
    tanggal = tanggal.text.strip() if tanggal != None else ''
    #print('tanggal', tanggal)
  except: tanggal = ''

  data_mobil = [merk, tipe, varian, harga, kondisi, tahun_kendaraan, kilometer, warna, mesin, transmisi, penumpang, pintu, dirakit, tenaga_puncak_hp, tenaga_putaran_puncak_nm, injeksi_langsung, aspirasi, bahan_bakar, panjang_mm, lebar_mm, tinggi_mm, jarak_roda_mm, seller_name, seller_status, seller_lokasi, judul, tanggal]
  list_mobil.append(data_mobil)
```

### Convert to Dataframe

```python
df_mobil = pd.DataFrame(list_mobil, columns=col_name)
```
